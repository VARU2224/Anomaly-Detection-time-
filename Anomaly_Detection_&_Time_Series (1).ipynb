**Question1**: What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

Answer:
Anomaly Detection is the process of identifying data points, patterns, or observations that deviate significantly from the expected or normal behavior in a dataset. These unusual observations, called anomalies or outliers, often indicate critical events such as fraud, system faults, network intrusions, or medical abnormalities.

**Types of Anomalies**
1. Point Anomalies
A point anomaly is a single data instance that is significantly different from the rest of the data.
Example:
In a credit card transaction dataset, a single transaction of ₹5,00,000 made by a user who typically spends under ₹2,000 is a point anomaly.
In temperature monitoring, a sudden reading of 95°C when all other values are around 30°C.


2. Contextual Anomalies
A contextual anomaly occurs when a data point is anomalous only within a specific context, such as time, location, or condition.

**Example:**
A temperature of 15°C may be normal in winter but anomalous in summer.
High website traffic at 2 PM may be normal, but the same traffic at 3 AM could indicate a potential cyberattack.


3. Collective Anomalies
A collective anomaly refers to a group of related data points that together form an unusual pattern, even though individual points may appear normal.
**Example:**
A sequence of small but repeated login failures may collectively indicate a brute-force attack.
In ECG signals, a series of normal-looking heartbeats that together form an abnormal rhythm pattern.


**Summary**

Point anomalies involve individual unusual data points.

Contextual anomalies depend on surrounding context like time or location.

Collective anomalies arise from abnormal patterns formed by groups of data points.

Anomaly detection is widely used in fraud detection, cybersecurity, healthcare monitoring, and fault detection systems.

**Question2**: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.

**Answer**:
1.Isolation Forest (IF)

Approach:
Isolation Forest is a tree-based anomaly detection algorithm that isolates anomalies instead of modeling normal data points. It works on the principle that anomalies are few and different, so they can be isolated with fewer random splits in decision trees.

**Key Characteristics:**

Does not rely on distance or density

Efficient on large and high-dimensional datasets

Works well even when data is not well clustered

Suitable Use Cases:

Fraud detection in financial transactions

Network intrusion detection

High-dimensional datasets (e.g., log data, sensor data)

When anomalies are rare and clearly different from normal data

2.DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Approach:
DBSCAN is a density-based clustering algorithm that identifies anomalies as points lying in low-density regions. Points that do not belong to any dense cluster are labeled as noise (outliers).

**Key Characteristics:**

Detects clusters of arbitrary shape

Requires tuning of eps and min_samples

Struggles with varying density and high dimensions

Suitable Use Cases:

Spatial or geographical data

Datasets with clear cluster structure

Noise detection in low-dimensional data

Image processing and geospatial analysis

3.Local Outlier Factor (LOF)

Approach:
LOF is a density-based local anomaly detection algorithm. It compares the local density of a point to the density of its neighbors. A point is considered anomalous if it has significantly lower density than its neighbors.

**Key Characteristics:**

Detects local anomalies

Sensitive to the choice of number of neighbors (k)

Computationally expensive for large datasets

Suitable Use Cases:

Detecting anomalies in non-uniform density datasets

Identifying local deviations rather than global ones

Intrusion detection, fault detection

When anomalies exist near clusters

**Comparative Summary**

Algorithm	Approach	Best For	Limitations
Isolation Forest	Tree-based isolation	Large, high-dimensional data	Less interpretable
DBSCAN	Density-based clustering	Spatial & clustered data	Sensitive to eps, struggles with varying density
LOF	Local density comparison	Local anomalies	Computationally expensive, parameter-sensitive

**Conclusion**

Use Isolation Forest for scalable, high-dimensional anomaly detection.

Use DBSCAN when anomalies are noise outside dense clusters.

Use LOF when anomalies are locally different within neighborhoods.

**Question3**: What are the key components of a Time Series? Explain each with one
example.
**Answer**:A Time Series is a sequence of observations recorded at regular time intervals (e.g., hourly, daily, monthly). Time series analysis focuses on identifying patterns in data over time to understand past behavior and forecast future values.

Key Components of a Time Series
1.Trend

The trend represents the long-term movement or direction of the data over time (upward, downward, or stable).

Example:

A company’s annual sales increasing steadily over several years show an upward trend.

2.Seasonality

Seasonality refers to regular, repeating patterns that occur at fixed intervals due to seasonal factors such as weather, holidays, or events.

Example:

Ice cream sales peaking every summer and dropping in winter show seasonality.

3.Cyclical Component

The cyclical component captures long-term fluctuations that occur over irregular periods, often influenced by economic or business cycles.

Example:

Economic growth and recession cycles affecting employment rates over multiple years.

4.Irregular (Random/Noise) Component

The irregular component consists of unpredictable, random variations that do not follow a pattern and cannot be explained by trend, seasonality, or cycles.

Example:

A sudden drop in stock prices due to unexpected political news.

Summary Table

Component	Meaning	Example
Trend	Long-term direction	Steady increase in yearly sales
Seasonality	Repeating short-term pattern	Higher ice cream sales in summer
Cyclical	Long-term economic cycles	Business booms and recessions
Irregular	Random fluctuations	Sudden market crash

Conclusion:
Understanding these components helps in effective forecasting, anomaly detection, and decision-making in real-world time series applications.

**Question4**: Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

**Answer**:
Stationarity in Time Series
A time series is said to be stationary if its statistical properties remain constant over time. In a stationary series, the mean, variance, and autocovariance do not change with time. Stationarity is a key assumption in many time series models such as ARIMA.

How to Test Stationarity

**1.Visual Inspection**

Plot the time series and check for trend or changing variance.

A stationary series fluctuates around a constant mean.

**2.Augmented Dickey–Fuller (ADF) Test**

Null Hypothesis (H₀): The series is non-stationary (has a unit root).

If p-value < 0.05, reject H₀ → series is stationary.

**3.KPSS Test**

Null Hypothesis (H₀): The series is stationary.

If p-value < 0.05, reject H₀ → series is non-stationary.

Using both ADF and KPSS provides a more reliable conclusion.

Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.
(Include your Python code and output in the code box below.)

Answer:Output Explanation

Scatter Plot

X-axis: Trip Distance

Y-axis: Fare Amount

Normal Points

Form a dense cluster

Represent typical NYC taxi trips with reasonable fares

Anomalies (Marked with ‘X’)

Appear far from the main cluster

Represent unusually long trips or excessively high fares

Correctly identified by Isolation Forest

**Conclusion**

Isolation Forest successfully detects anomalous NYC taxi trips and visually separates them from normal rides.
This approach is commonly used in fraud detection, pricing anomalies, and transportation analytics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# -----------------------------
# Create a NYC Taxi Fare–like dataset
# -----------------------------
np.random.seed(42)

# Normal taxi trips
trip_distance = np.random.normal(loc=5, scale=2, size=400)
fare_amount = trip_distance * 2.5 + np.random.normal(0, 2, size=400)

# Anomalous trips (very high fares/distances)
anomaly_distance = np.random.uniform(20, 40, size=20)
anomaly_fare = anomaly_distance * 4 + np.random.uniform(20, 50, size=20)

# Combine data
distance = np.concatenate([trip_distance, anomaly_distance])
fare = np.concatenate([fare_amount, anomaly_fare])

data = pd.DataFrame({
    "trip_distance": distance,
    "fare_amount": fare
})

# -----------------------------
# Apply Isolation Forest
# -----------------------------
iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.05,
    random_state=42
)

data["anomaly"] = iso_forest.fit_predict(data)

# Convert labels: -1 = anomaly, 1 = normal
data["anomaly"] = data["anomaly"].map({1: 0, -1: 1})

# -----------------------------
# Visualization
# -----------------------------
plt.figure()
plt.scatter(
    data[data["anomaly"] == 0]["trip_distance"],
    data[data["anomaly"] == 0]["fare_amount"],
    label="Normal",
    alpha=0.6
)

plt.scatter(
    data[data["anomaly"] == 1]["trip_distance"],
    data[data["anomaly"] == 1]["fare_amount"],
    label="Anomaly",
    marker="x"
)

plt.xlabel("Trip Distance (miles)")
plt.ylabel("Fare Amount ($)")
plt.title("Isolation Forest Anomaly Detection (NYC Taxi Fare)")
plt.legend()
plt.show()
```

Question 8: Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.
(Include your Python code and output in the code box below.)

Answer:SARIMA Was Used

Handles trend via differencing (d = 1)

Handles seasonality via seasonal differencing (D = 1, s = 12)

Ideal for monthly airline passenger data

**Output Explanation**

*Observed Series*

*Displays historical monthly airline passenger counts*

*Shows a clear upward trend and strong yearly seasonality*

*Forecast Line*

*Predicts passenger numbers for the next 12 months*

*Continues the upward trend with seasonal fluctuations*

*Confidence Interval (Shaded Region)*

*Represents uncertainty in the forecast*

*Widens slightly over time, which is expected*

**Conclusion**

The SARIMA model successfully captures both trend and seasonality in the AirPassengers dataset and provides realistic forecasts for the next year.

This approach is widely used in time series forecasting problems involving seasonal data.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# -----------------------------
# Load AirPassengers dataset
# -----------------------------
data = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
)

# Convert Month to datetime and set index
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)
data.columns = ['Passengers']

# -----------------------------
# Train SARIMA model
# SARIMA(p,d,q)(P,D,Q,s)
# -----------------------------
model = SARIMAX(
    data['Passengers'],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model.fit()

# -----------------------------
# Forecast next 12 months
# -----------------------------
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# -----------------------------
# Visualization
# -----------------------------
plt.figure()
plt.plot(data, label='Observed')
plt.plot(forecast_mean, label='Forecast')
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    alpha=0.3
)

plt.title('SARIMA Forecast: AirPassengers (Next 12 Months)')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

Answer:**Output Explanation**

Scatter Plot

Each point represents a data observation with two numerical features.

Normal Points

Anomalies (Marked with ‘X’)

**Conclusion**

Local Outlier Factor successfully identifies anomalous points by comparing local neighborhood density, and the visualization clearly separates anomalies from normal data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# -----------------------------
# Create a numerical dataset
# -----------------------------
np.random.seed(42)

# Normal data points
x_normal = np.random.normal(0, 1, 300)
y_normal = np.random.normal(0, 1, 300)

# Anomalous data points
x_anomaly = np.random.uniform(5, 8, 20)
y_anomaly = np.random.uniform(5, 8, 20)

# Combine data
X = np.column_stack((
    np.concatenate([x_normal, x_anomaly]),
    np.concatenate([y_normal, y_anomaly])
))

data = pd.DataFrame(X, columns=["Feature1", "Feature2"])

# -----------------------------
# Apply Local Outlier Factor
# -----------------------------
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.06)
labels = lof.fit_predict(data)

# Convert labels: -1 = anomaly, 1 = normal
data["anomaly"] = labels

# -----------------------------
# Visualization
# -----------------------------
plt.figure()
plt.scatter(
    data[data["anomaly"] == 1]["Feature1"],
    data[data["anomaly"] == 1]["Feature2"],
    label="Normal"
)

plt.scatter(
    data[data["anomaly"] == -1]["Feature1"],
    data[data["anomaly"] == -1]["Feature2"],
    marker="x",
    label="Anomaly"
)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# -----------------------------
# Simulated 15-min Energy Data
# -----------------------------
np.random.seed(42)

min = pd.date_range("2024-01-01", periods=500, freq="15T")
energy = 100 + np.random.normal(0, 5, 500)
energy[450:455] += 40   # spike anomaly
energy[480:485] -= 35   # drop anomaly

temperature = 25 + np.random.normal(0, 2, 500)

data = pd.DataFrame({
    "energy": energy,
    "temperature": temperature
}, index=min)

# -----------------------------
# Anomaly Detection
# -----------------------------
iso = IsolationForest(
    contamination=0.03,
    random_state=42
)
data["anomaly"] = iso.fit_predict(data[["energy", "temperature"]])

# -----------------------------
# Forecasting with SARIMAX
# -----------------------------
model = SARIMAX(
    data["energy"],
    exog=data[["temperature"]],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 96)  # 96 = daily seasonality (15 min data)
)

results = model.fit(disp=False)
forecast = results.get_forecast(
    steps=32,
    exog=data[["temperature"]].iloc[-32:]
)
forecast_mean = forecast.predicted_mean

# -----------------------------
# Visualization
# -----------------------------
plt.figure()
plt.plot(data.index, data["energy"], label="Energy Usage")
plt.scatter(
    data[data["anomaly"] == -1].index,
    data[data["anomaly"] == -1]["energy"],
    marker="x",
    label="Anomaly"
)
plt.plot(forecast_mean.index, forecast_mean, label="Forecast")
plt.legend()
plt.title("Energy Demand Forecasting & Anomaly Detection")
plt.xlabel("Time")
plt.ylabel("Energy Usage")
plt.show()
```
